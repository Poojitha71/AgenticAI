```python
num = int(input(" enter a number"))
if num>0:
    print("number is positive")
    if num%2 == 0:
        print("num is even")
    else:
        print("num is odd")
elif num==0:
    print("num is 0")
else:
    print("num is negative")
```

```python
year = int(input("enter a year:"))
if year %4==0:
    if year%100==0:
        if year%400==0:
            print(year,"year is  a leap year")
        else:
            print(year, "not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"not a leap year")
```

```python
class Person:
    def __init__(self,name,age):
        self.__name = name ## private access modifier
        self.__age = age ## private access modifier
        
    ## getter method for name variable
    def get_name(self):
        return self.__name
    
    ## setter method for name variable
    def set_name(self,name):
        self.__name = name
        
    ## getter method for age variable
    def get_age(self):
        return self.__age
    
    ## setter method for age variable
    def set_age(self,age):
        if age>0:
            self.__age = age
        else:
            print("age doent accept negative value")
        
        
person = Person("pooj",23)

print(person.get_name())
print(person.get_age())

person.set_age(24)
print(person.get_age())

person.set_age(-24)
```

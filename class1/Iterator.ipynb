```python
my_list = [1,2,3,4,5]

for i in my_list:
    print(i)
    
type(my_list)
```

```python
## iterator 

iterator = iter(my_list)
type(iterator)
```

```python
next(iterator)
```

```python
my_list = [1,2,3,4,5]
iterator = iter(my_list)
```

```python

try :
    print(next(iterator))
    
except StopIteration:
    print("there are no elements in the list to iterate")
    
```

```python
## string iterator
myString = "hello!"
iterator = iter(myString)
```

```python
try:
    print(next(iterator))
except StopIteration:
    print("there are no other elements in the string")
```

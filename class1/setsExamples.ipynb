```python
my_set ={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
## basic set opeartions
# adding elements
my_set.add(7)
print(my_set)

# removing elements
my_set.remove(7)
print(my_set)

# pop elements
my_set.pop()
print(my_set)

my_set.clear()
print(my_set)

```

```python
## set membership test
my_set ={1,2,3,4,5}
print(3 in my_set)
print(6 in my_set)
```

```python
## mathematical operation

set1={1,2,3,4,5,6}
set2={7,8,9,2,3,4}

# union
union_set=set1.union(set2)
print(f"union-set : {union_set}")

#intersection
intersection_set = set1.intersection(set2)
print(f"intersection set : {intersection_set}")

#intersection_update
intersection_update = set1.intersection_update(set2)
print(f"intersection-update : {set1}")

# difference 
set1={1,2,3,4,5,6}
set2={7,8,9,2,3,4}
difference = set1.difference(set2)
print(f"difference: {difference}")

#symmentic difference
print(set1.symmetric_difference(set2))

```

```python
## set methods
set1={1,2,3,4,5,6}
set2={7,8,9,2,3,4}

# is subset
print(set1.issubset(set2))

# is superset
print(set1.issuperset(set2))
```

```python
## list to set
my_list = [1,2,3,4,5]
set_list = set(my_list)
print(set_list)
print(type(set_list))
```

```python
## counting unique words in text
text = " hi my name is poojitha . today i am learning set examples"

words = text.split()
print(type(words))
print(words)
# convert list of words to set 
unique_words =set(words)
print(unique_words)
print(len(unique_words))
```
